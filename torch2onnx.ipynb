```python
import torch

from superpoint.model import SuperPointNet

weights_path = "/home/plus/tensorrt/tensorrt/superpoint/superpoint_v1.pth"
model = SuperPointNet()
model.load_state_dict(torch.load(weights_path))
model = model.cuda()
x = torch.rand(1, 1, 800, 1280).cuda()
torch.onnx.export(model,
                    x,
                    'superpoint.onnx',
                    opset_version = 11,
                    verbose = True,
                    input_names = ['input'],
                    output_names = ['output'])

```

```python
import torch
 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
weights_path = "/home/plus/Work/tensorrt/superpoint/superpoint_v1.pth"
 
def model_converter():
    model = SuperPointNet()
    model.load_state_dict(torch.load(weights_path))
    model.to(device)  # 这里保存的是完整模型
    model.eval()
 
    dummy_input = torch.randn(1, 1, 800, 1280, device=device)
    input_names = ['input']
    output_names = ['output']
    torch.onnx.export(model, dummy_input, 'superpoint.onnx',
                      export_params=True,
                      verbose=True,
                      input_names=input_names,
                      output_names=output_names)
 
model_converter()
```
